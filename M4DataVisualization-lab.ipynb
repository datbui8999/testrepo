## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
import pandas as pd
QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
plt.hist(df['ConvertedComp'], bins=50, edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Count')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""
df=pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(6,8))
sns.boxplot(data=df)
plt.title('Box plot of Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY= """
SELECT Age, WorkWeekHrs
FROM master
"""
df= pd.read_sql_query(QUERY,conn)
sns.scatterplot(data=df,x = 'Age', y= 'WorkWeekHrs')
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY= """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df= pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(15,15))
sns.scatterplot(data=df,x = 'WorkWeekHrs', y= 'CodeRevHrs', size='Age',alpha = 0.6, sizes=(1, 500))
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY,conn)
data=df['count']
labels = df['DatabaseDesireNextYear']
fig, ax=plt.subplots()
ax=plt.pie(data=df, x=data, labels=labels, autopct='%1.1f%%', explode = (0.1, 0, 0, 0, 0), shadow = True)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)
work_week_median = df['WorkWeekHrs'].median()
code_rev_median = df['CodeRevHrs'].median()
plt.figure(figsize=(8, 8))
sns.barplot(x=['Age 30 - 35'], y=[work_week_median], color='blue', label='WorkWeekHrs')
sns.barplot(x=['Age 30 - 35'], y=[code_rev_median], color='orange', bottom=[work_week_median], label='CodeRevHrs')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age 30 - 35')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY="""
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df=pd.read_sql_query(QUERY, conn)
medians= df.groupby('Age')['ConvertedComp'].median()
data = medians.reset_index(name='MedConvertedComp')
plt.rcParams['figure.figsize'] = [8,8]
sns.lineplot(data=data, x='Age', y = 'MedConvertedComp')
plt.show()
```

Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency
FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY,conn)
plt.rcParams['figure.figsize'] = [14, 8]
sns.barplot(data=df, y='MainBranch',x='Frequency', orient='h')
plt.show()
```
